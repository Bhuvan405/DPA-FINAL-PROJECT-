```python
# Import all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, LabelEncoder, PolynomialFeatures
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression, LassoCV
from sklearn.svm import SVC
from sklearn.feature_selection import SelectFromModel, RFE, mutual_info_classif
from sklearn.metrics import (
    classification_report, accuracy_score, roc_auc_score, roc_curve,
    silhouette_score, confusion_matrix, ConfusionMatrixDisplay,
    calinski_harabasz_score, davies_bouldin_score
)
from sklearn.neural_network import MLPRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from scipy.stats import chi2_contingency

import warnings
warnings.filterwarnings('ignore')

def analyze_missing_values(data):
    """Analyze missing values in the dataset."""
    missing_summary = pd.DataFrame({
        'Missing Values': data.isnull().sum(),
        'Percentage': (data.isnull().sum() / len(data)) * 100
    })
    return missing_summary[missing_summary['Missing Values'] > 0]

def detect_outliers(data, columns):
    """Detect outliers using IQR method."""
    outliers = {}
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[column] = {
            'count': len(data[(data[column] < lower_bound) | (data[column] > upper_bound)]),
            'percentage': len(data[(data[column] < lower_bound) | (data[column] > upper_bound)]) / len(data) * 100
        }
    return pd.DataFrame(outliers).T

def load_and_prepare_data(file_path):
    """Load and prepare the data with enhanced analysis."""
    print("\n1. Data Loading and Initial Analysis")

    # Load data
    data = pd.read_csv(file_path)
    print("Initial data shape:", data.shape)

    # Analyze missing values
    missing_data = analyze_missing_values(data)
    if not missing_data.empty:
        print("\nMissing Value Analysis:")
        print(missing_data)

    # Remove client number if present
    if 'CLIENTNUM' in data.columns:
        data = data.drop('CLIENTNUM', axis=1)

    # Analyze numerical features
    numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

    # Detect outliers in numerical columns
    print("\nOutlier Analysis:")
    outliers_summary = detect_outliers(data, numerical_columns)
    print(outliers_summary)

    # Encode categorical variables
    le = LabelEncoder()
    categorical_columns = data.select_dtypes(include=['object']).columns
    for col in categorical_columns:
        data[col] = le.fit_transform(data[col])

    # Split features and target
    y = data['Attrition_Flag']
    X = data.drop('Attrition_Flag', axis=1)

    # Scale features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    return X, X_scaled, y, data, X.columns, numerical_columns, categorical_columns

def explore_data(data, X_scaled, feature_names, numerical_columns, categorical_columns):
    """Enhanced exploratory data analysis."""
    print("\n2. Enhanced Data Exploration")

    # Numerical features distribution
    plt.figure(figsize=(15, 5 * ((len(numerical_columns) - 1) // 3 + 1)))
    for i, col in enumerate(numerical_columns, 1):
        if col != 'Attrition_Flag':
            plt.subplot((len(numerical_columns) - 1) // 3 + 1, 3, i)
            sns.histplot(data=data, x=col, hue='Attrition_Flag', multiple="stack")
            plt.title(f'{col} Distribution')
    plt.tight_layout()
    plt.show()

    # Correlation analysis
    plt.figure(figsize=(12, 8))
    correlation = data.corr()
    sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')
    plt.title('Feature Correlation Matrix')
    plt.tight_layout()
    plt.show()

    # Identify highly correlated features
    high_correlation = np.where(np.abs(correlation) > 0.7)
    high_correlation = [(correlation.index[x], correlation.columns[y], correlation.iloc[x, y])
                       for x, y in zip(*high_correlation) if x != y]
    if high_correlation:
        print("\nHighly correlated features (|correlation| > 0.7):")
        for feat1, feat2, corr in high_correlation:
            print(f"{feat1} - {feat2}: {corr:.3f}")

    # Categorical features analysis
    for col in categorical_columns:
        if col != 'Attrition_Flag':
            contingency = pd.crosstab(data[col], data['Attrition_Flag'])
            chi2, p_value = chi2_contingency(contingency)[:2]
            print(f"\nChi-square test for {col}:")
            print(f"Chi-square statistic: {chi2:.2f}")
            print(f"p-value: {p_value:.4f}")

    # Feature independence analysis
    mi_scores = mutual_info_classif(X_scaled, data['Attrition_Flag'])
    mi_df = pd.DataFrame({'Feature': feature_names, 'MI_Score': mi_scores})
    mi_df = mi_df.sort_values('MI_Score', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(data=mi_df.head(10), x='MI_Score', y='Feature')
    plt.title('Top 10 Features by Mutual Information Score')
    plt.tight_layout()
    plt.show()

    return mi_df

def dimensionality_reduction(X_scaled, y, feature_names):
    """Enhanced dimensionality reduction analysis."""
    print("\n3. Enhanced Dimensionality Reduction Analysis")

    # PCA with component interpretation
    pca = PCA()
    pca_result = pca.fit_transform(X_scaled)

    # Analyze PCA components
    components_df = pd.DataFrame(
        pca.components_,
        columns=feature_names
    )

    plt.figure(figsize=(12, 6))
    sns.heatmap(components_df.iloc[:3], annot=True, fmt='.2f', cmap='coolwarm')
    plt.title('PCA Components Interpretation')
    plt.tight_layout()
    plt.show()

    # Explained variance analysis
    plt.figure(figsize=(10, 5))
    cumsum = np.cumsum(pca.explained_variance_ratio_)
    plt.plot(range(1, len(cumsum) + 1), cumsum, 'bo-')
    plt.axhline(y=0.95, color='r', linestyle='--', label='95% Explained Variance')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance Ratio')
    plt.title('PCA Explained Variance Analysis')
    plt.legend()
    plt.grid(True)
    plt.show()

    # t-SNE with perplexity tuning
    perplexities = [5, 30, 50]
    plt.figure(figsize=(15, 5))
    tsne_result = None
    for i, perp in enumerate(perplexities, 1):
        tsne = TSNE(n_components=2, random_state=42, perplexity=perp)
        tsne_temp = tsne.fit_transform(X_scaled)
        if perp == 30:  # Save the middle perplexity result
            tsne_result = tsne_temp
        plt.subplot(1, 3, i)
        plt.scatter(tsne_temp[:, 0], tsne_temp[:, 1], c=y, cmap='viridis')
        plt.title(f't-SNE (perplexity={perp})')
    plt.tight_layout()
    plt.show()

    # UMAP visualization
    reducer = umap.UMAP(random_state=42)
    umap_result = reducer.fit_transform(X_scaled)

    plt.figure(figsize=(10, 5))
    plt.scatter(umap_result[:, 0], umap_result[:, 1], c=y, cmap='viridis')
    plt.title('UMAP Projection')
    plt.colorbar(label='Target')
    plt.tight_layout()
    plt.show()

    return pca_result, tsne_result, umap_result
def clustering_analysis(X_scaled, y):
    """Enhanced clustering analysis with error handling."""
    print("\n4. Enhanced Clustering Analysis")

    # K-means analysis
    k_range = range(2, 11)
    metrics = {
        'inertia': [],
        'silhouette': [],
        'calinski_harabasz': [],
        'davies_bouldin': []
    }

    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X_scaled)

        metrics['inertia'].append(kmeans.inertia_)
        metrics['silhouette'].append(silhouette_score(X_scaled, labels))
        metrics['calinski_harabasz'].append(calinski_harabasz_score(X_scaled, labels))
        metrics['davies_bouldin'].append(davies_bouldin_score(X_scaled, labels))

    # Plot clustering metrics
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    axes[0, 0].plot(k_range, metrics['inertia'], 'bo-')
    axes[0, 0].set_title('Elbow Method')
    axes[0, 0].set_xlabel('Number of Clusters')
    axes[0, 0].set_ylabel('Inertia')

    axes[0, 1].plot(k_range, metrics['silhouette'], 'ro-')
    axes[0, 1].set_title('Silhouette Score')
    axes[0, 1].set_xlabel('Number of Clusters')
    axes[0, 1].set_ylabel('Score')

    axes[1, 0].plot(k_range, metrics['calinski_harabasz'], 'go-')
    axes[1, 0].set_title('Calinski-Harabasz Score')
    axes[1, 0].set_xlabel('Number of Clusters')
    axes[1, 0].set_ylabel('Score')

    axes[1, 1].plot(k_range, metrics['davies_bouldin'], 'mo-')
    axes[1, 1].set_title('Davies-Bouldin Score')
    axes[1, 1].set_xlabel('Number of Clusters')
    axes[1, 1].set_ylabel('Score')

    plt.tight_layout()
    plt.show()

    # DBSCAN parameter tuning
    eps_range = np.linspace(0.1, 1.0, 10)
    min_samples_range = [5, 10, 15]

    dbscan_results = []

    for eps in eps_range:
        for min_samples in min_samples_range:
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            labels = dbscan.fit_predict(X_scaled)

            n_clusters = len(np.unique(labels[labels >= 0]))
            n_noise = np.sum(labels == -1)

            try:
                if n_clusters > 1:
                    silhouette = silhouette_score(X_scaled, labels)
                    dbscan_results.append({
                        'eps': eps,
                        'min_samples': min_samples,
                        'n_clusters': n_clusters,
                        'n_noise': n_noise,
                        'n_noise_percent': n_noise / len(labels) * 100,
                        'silhouette_score': silhouette
                    })
            except Exception as e:
                print(f"Error calculating metrics for eps={eps}, min_samples={min_samples}: {str(e)}")
                continue

    if dbscan_results:
        dbscan_df = pd.DataFrame(dbscan_results)
        print("\nTop DBSCAN configurations:")
        print(dbscan_df.sort_values('silhouette_score', ascending=False).head())

        best_config = dbscan_df.loc[dbscan_df['silhouette_score'].idxmax()]
        best_dbscan = DBSCAN(eps=best_config['eps'], min_samples=int(best_config['min_samples']))
        best_labels = best_dbscan.fit_predict(X_scaled)

        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=best_labels, cmap='viridis')
        plt.title(f'Best DBSCAN Clustering\neps={best_config["eps"]:.2f}, min_samples={int(best_config["min_samples"])}')

        plt.subplot(1, 2, 2)
        eps_scores = dbscan_df.groupby('eps')['silhouette_score'].mean()
        plt.plot(eps_scores.index, eps_scores.values, 'bo-')
        plt.xlabel('eps')
        plt.ylabel('Average Silhouette Score')
        plt.title('DBSCAN Parameter Sensitivity')
        plt.tight_layout()
        plt.show()
    else:
        print("\nNo valid DBSCAN configurations found. Try adjusting the parameter ranges.")
        dbscan_df = pd.DataFrame()

    return metrics, dbscan_df

def feature_engineering(X_scaled, feature_names):
    """Perform feature engineering."""
    print("\n5. Feature Engineering")

    # Create polynomial features
    poly = PolynomialFeatures(degree=2, include_bias=False)
    X_poly = poly.fit_transform(X_scaled)

    # Get feature names for polynomial features
    poly_features = []
    powers = poly.powers_[1:]
    for powers_tuple in powers:
        feature = ''
        for feat, power in zip(feature_names, powers_tuple):
            if power > 0:
                feature += f"{feat}^{power}_"
        poly_features.append(feature[:-1])

    return X_poly, poly_features

def evaluate_cv_results(cv_scores):
    """Analyze cross-validation results."""
    mean_score = np.mean(cv_scores)
    std_score = np.std(cv_scores)
    print(f"Mean CV Score: {mean_score:.3f} (+/- {std_score * 2:.3f})")
    return mean_score, std_score

def train_evaluate_model(model, X_train, X_test, y_train, y_test, model_name, param_grid):
    """Train and evaluate a single model."""
    # Cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc')
    print(f"\n{model_name} Cross-validation Results:")
    mean_cv, std_cv = evaluate_cv_results(cv_scores)

    # Grid search
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='roc_auc', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Predictions
    y_pred = grid_search.predict(X_test)
    y_prob = grid_search.predict_proba(X_test)[:, 1]

    # Performance metrics
    print(f"\n{model_name} Best Parameters:", grid_search.best_params_)
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = roc_auc_score(y_test, y_prob)

    # Confusion matrix
    plt.figure(figsize=(8, 6))
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()

    return {
        'model': grid_search,
        'cv_scores': cv_scores,
        'y_pred': y_pred,
        'y_prob': y_prob,
        'fpr': fpr,
        'tpr': tpr,
        'roc_auc': roc_auc
    }

def compare_models(model_results):
    """Compare performance of multiple models."""
    print("\n6. Model Comparison")

    # Plot ROC curves
    plt.figure(figsize=(10, 8))
    for name, results in model_results.items():
        plt.plot(results['fpr'], results['tpr'], label=f'{name} (AUC = {results["roc_auc"]:.3f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves Comparison')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Compare cross-validation scores
    cv_means = [np.mean(results['cv_scores']) for results in model_results.values()]
    cv_stds = [np.std(results['cv_scores']) for results in model_results.values()]

    plt.figure(figsize=(10, 6))
    plt.bar(model_results.keys(), cv_means, yerr=cv_stds)
    plt.title('Cross-validation Scores Comparison')
    plt.ylabel('Mean ROC-AUC Score')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def train_models(X_scaled, y, X_poly=None):
    """Train and evaluate multiple models."""
    print("\n7. Model Training and Evaluation")

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y
    )

    # Handle class imbalance
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    # Baseline model (majority class)
    majority_pred = np.full(len(y_test), stats.mode(y_train)[0])
    print("\nBaseline (Majority Class) Performance:")
    print(classification_report(y_test, majority_pred))

    # Define models and their parameter grids
    models = {
        'Logistic Regression': (
            LogisticRegression(random_state=42),
            {
                'C': [0.001, 0.01, 0.1, 1, 10],
                'penalty': ['l1', 'l2'],
                'solver': ['liblinear']
            }
        ),
        'Random Forest': (
            RandomForestClassifier(random_state=42),
            {
                'n_estimators': [100, 200],
                'max_depth': [None, 10, 20],
                'min_samples_split': [2, 5],
                'class_weight': ['balanced', 'balanced_subsample']
            }
        ),
        'Gradient Boosting': (
            GradientBoostingClassifier(random_state=42),
            {
                'n_estimators': [100, 200],
                'learning_rate': [0.01, 0.1],
                'max_depth': [3, 5]
            }
        ),
        'SVM': (
            SVC(random_state=42, probability=True),
            {
                'C': [0.1, 1, 10],
                'kernel': ['rbf', 'linear'],
                'class_weight': ['balanced']
            }
        )
    }

    # Train and evaluate models
    model_results = {}
    for name, (model, param_grid) in models.items():
        print(f"\nTraining {name}...")
        model_results[name] = train_evaluate_model(
            model, X_train_resampled, X_test, y_train_resampled, y_test,
            name, param_grid
        )

    # Compare models
    compare_models(model_results)

    # If polynomial features are provided, train models with them
    if X_poly is not None:
        print("\nTraining models with polynomial features...")
        X_poly_train, X_poly_test = train_test_split(
            X_poly, test_size=0.2, random_state=42
        )
        X_poly_train_resampled, y_poly_train_resampled = smote.fit_resample(X_poly_train, y_train)

        poly_model_results = {}
        for name, (model, param_grid) in models.items():
            print(f"\nTraining {name} with polynomial features...")
            poly_model_results[f"{name}_Poly"] = train_evaluate_model(
                model, X_poly_train_resampled, X_poly_test, y_poly_train_resampled, y_test,
                f"{name} (Polynomial)", param_grid
            )

        # Compare polynomial models
        compare_models(poly_model_results)
        model_results.update(poly_model_results)

    return model_results

def main():
    """Main function to run the complete analysis pipeline."""
    # Load and prepare data
    X, X_scaled, y, data, feature_names, numerical_columns, categorical_columns = load_and_prepare_data('BankChurners.csv')

    # Perform all analyses
    mi_df = explore_data(data, X_scaled, feature_names, numerical_columns, categorical_columns)
    pca_result, tsne_result, umap_result = dimensionality_reduction(X_scaled, y, feature_names)
    clustering_metrics, dbscan_results = clustering_analysis(X_scaled, y)

    # Feature engineering
    X_poly, poly_features = feature_engineering(X_scaled, feature_names)

    # Train models with original and engineered features
    model_results = train_models(X_scaled, y, X_poly)

    # Print final summary
    print("\nAnalysis Complete!")
    print("\nKey Findings:")
    print("1. Top features by mutual information:")
    print(mi_df.head().to_string())
    print("\n2. Best performing model:")
    best_model = max(model_results.items(), key=lambda x: x[1]['roc_auc'])
    print(f"{best_model[0]} (ROC-AUC: {best_model[1]['roc_auc']:.3f})")

    return {
        'data': data,
        'X': X,
        'X_scaled': X_scaled,
        'y': y,
        'pca_result': pca_result,
        'tsne_result': tsne_result,
        'umap_result': umap_result,
        'clustering_metrics': clustering_metrics,
        'dbscan_results': dbscan_results,
        'model_results': model_results,
        'feature_importance': mi_df
    }

if __name__ == "__main__":
    results = main()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Load dataset
file_path = "/content/BankChurners.csv"
data = pd.read_csv(file_path)

# Define target variable
data['Attrition_Flag_Binary'] = data['Attrition_Flag'].apply(lambda x: 1 if x == 'Attrited Customer' else 0)
y_no_leakage = data['Attrition_Flag_Binary']

# Drop leakage features
leakage_features = [
    'Attrition_Flag_Binary',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'
]
X_no_leakage = data.drop(columns=leakage_features + ['Attrition_Flag'])

# Identify categorical features
categorical_features = X_no_leakage.select_dtypes(include=['object']).columns

# One-hot encode categorical features
encoder = OneHotEncoder(sparse_output=False, drop='first')  # Use sparse_output instead of sparse
encoded_cats = encoder.fit_transform(X_no_leakage[categorical_features])
encoded_cats_df = pd.DataFrame(encoded_cats, columns=encoder.get_feature_names_out(categorical_features))

# Combine numerical and encoded categorical features
X_no_leakage_combined = pd.concat(
    [X_no_leakage.drop(columns=categorical_features).reset_index(drop=True), encoded_cats_df.reset_index(drop=True)],
    axis=1
)

# Scale the features
scaler = StandardScaler()
X_scaled_no_leakage = scaler.fit_transform(X_no_leakage_combined)

# Initialize logistic regression and stratified k-fold
logistic_model_no_leakage = LogisticRegression(random_state=42, max_iter=1000)
stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform stratified k-fold cross-validation
metrics_no_leakage = []

for fold, (train_idx, val_idx) in enumerate(stratified_kfold.split(X_scaled_no_leakage, y_no_leakage)):
    # Split data
    X_train, X_val = X_scaled_no_leakage[train_idx], X_scaled_no_leakage[val_idx]
    y_train, y_val = y_no_leakage.iloc[train_idx], y_no_leakage.iloc[val_idx]

    # Train logistic regression model
    logistic_model_no_leakage.fit(X_train, y_train)

    # Predict and evaluate
    y_pred = logistic_model_no_leakage.predict(X_val)
    fold_metrics = {
        "Fold": fold + 1,
        "Accuracy": accuracy_score(y_val, y_pred),
        "Precision": precision_score(y_val, y_pred),
        "Recall": recall_score(y_val, y_pred),
        "F1-Score": f1_score(y_val, y_pred),
    }
    metrics_no_leakage.append(fold_metrics)

# Convert metrics to DataFrame
metrics_no_leakage_df = pd.DataFrame(metrics_no_leakage)
print("Performance Metrics for Logistic Regression (No Leakage):")
print(metrics_no_leakage_df)

# Display average metrics
average_metrics_no_leakage = metrics_no_leakage_df.mean()
print("\nAverage Metrics (No Leakage):")
print(average_metrics_no_leakage)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define parameter grid for logistic regression with regularization
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength (inverse of alpha)
    'penalty': ['l1', 'l2'],  # L1 = Lasso, L2 = Ridge
    'solver': ['liblinear']  # Solver supporting L1 and L2
}

# Set up grid search with stratified k-fold cross-validation
grid_search = GridSearchCV(
    estimator=LogisticRegression(random_state=42, max_iter=1000),
    param_grid=param_grid,
    scoring='f1',  # Optimize for F1-Score
    cv=stratified_kfold,
    n_jobs=-1
)

# Fit grid search on the dataset without leakage
grid_search.fit(X_scaled_no_leakage, y_no_leakage)

# Extract best parameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model using stratified k-fold cross-validation
metrics_regularized = []

for fold, (train_idx, val_idx) in enumerate(stratified_kfold.split(X_scaled_no_leakage, y_no_leakage)):
    X_train, X_val = X_scaled_no_leakage[train_idx], X_scaled_no_leakage[val_idx]
    y_train, y_val = y_no_leakage.iloc[train_idx], y_no_leakage.iloc[val_idx]

    # Train the best model
    best_model.fit(X_train, y_train)

    # Predict and evaluate
    y_pred = best_model.predict(X_val)
    fold_metrics = {
        "Fold": fold + 1,
        "Accuracy": accuracy_score(y_val, y_pred),
        "Precision": precision_score(y_val, y_pred),
        "Recall": recall_score(y_val, y_pred),
        "F1-Score": f1_score(y_val, y_pred),
    }
    metrics_regularized.append(fold_metrics)

# Convert metrics to DataFrame
metrics_regularized_df = pd.DataFrame(metrics_regularized)
print("Performance Metrics for Regularized Logistic Regression:")
print(metrics_regularized_df)

# Display average metrics
average_metrics_regularized = metrics_regularized_df.mean()
print("\nAverage Metrics (Regularized):")
print(average_metrics_regularized)

# Display best parameters
print("\nBest Parameters from Grid Search:")
print(best_params)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Load dataset
file_path = "/content/BankChurners.csv"
data = pd.read_csv(file_path)

# Define target variable
data['Attrition_Flag_Binary'] = data['Attrition_Flag'].apply(lambda x: 1 if x == 'Attrited Customer' else 0)
y_no_leakage = data['Attrition_Flag_Binary']

# Drop leakage features
leakage_features = [
    'Attrition_Flag_Binary',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'
]
X_no_leakage = data.drop(columns=leakage_features + ['Attrition_Flag'])

# Identify categorical features
categorical_features = X_no_leakage.select_dtypes(include=['object']).columns

# One-hot encode categorical features
encoder = OneHotEncoder(sparse_output=False, drop='first')  # Updated argument
encoded_cats = encoder.fit_transform(X_no_leakage[categorical_features])
encoded_cats_df = pd.DataFrame(encoded_cats, columns=encoder.get_feature_names_out(categorical_features))

# Combine numerical and encoded categorical features
X_no_leakage_combined = pd.concat(
    [X_no_leakage.drop(columns=categorical_features).reset_index(drop=True), encoded_cats_df.reset_index(drop=True)],
    axis=1
)

# Scale the features
scaler = StandardScaler()
X_scaled_no_leakage = scaler.fit_transform(X_no_leakage_combined)

# Initialize stratified k-fold
stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42, n_estimators=100, max_depth=None, n_jobs=-1)

# Evaluate Random Forest using stratified k-fold cross-validation
metrics_rf = []

for fold, (train_idx, val_idx) in enumerate(stratified_kfold.split(X_scaled_no_leakage, y_no_leakage)):
    # Split data into training and validation sets
    X_train, X_val = X_scaled_no_leakage[train_idx], X_scaled_no_leakage[val_idx]
    y_train, y_val = y_no_leakage.iloc[train_idx], y_no_leakage.iloc[val_idx]

    # Train the Random Forest model
    rf_model.fit(X_train, y_train)

    # Predict and evaluate
    y_pred = rf_model.predict(X_val)
    fold_metrics = {
        "Fold": fold + 1,
        "Accuracy": accuracy_score(y_val, y_pred),
        "Precision": precision_score(y_val, y_pred),
        "Recall": recall_score(y_val, y_pred),
        "F1-Score": f1_score(y_val, y_pred),
    }
    metrics_rf.append(fold_metrics)

# Convert metrics to DataFrame
metrics_rf_df = pd.DataFrame(metrics_rf)
print("Performance Metrics for Random Forest:")
print(metrics_rf_df)

# Display average metrics
average_metrics_rf = metrics_rf_df.mean()
print("\nAverage Metrics (Random Forest):")
print(average_metrics_rf)

```
